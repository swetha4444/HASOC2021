```python
import pandas as pd
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

## SUB-TASK A (DistilBERT)

```python
df_test1 = pd.read_csv('../../Submissions/Independent/BERT_A.csv')
df_result1 = pd.read_csv('1A_English_GOLD.csv')
```

```python
cm = confusion_matrix(df_result1['label'], df_test1['label'], labels=['HOF','NOT'])
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=['HOF','NOT'])
disp.plot() 
```
