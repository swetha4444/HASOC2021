```python
import pandas as pd
```

```python
df_ans = pd.read_csv("../preprocess_test_data.csv", index_col = 0)
df_ans['text_clean'] = df_ans['text_clean'].fillna('neutral')
df_ans.head()
```

```python
import sklearn
sklearn.__version__
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidconverter = TfidfVectorizer(max_features=2000, min_df= 2)
X_test = tfidconverter.fit_transform(df_ans['text_clean']).toarray()
```

```python
print(len(X_test[1]))
```

```python
import pickle
dt1 = pickle.load(open('finalized_svc_model1.sav', 'rb'))
dt2 = pickle.load(open('finalized_rfc_model1.sav', 'rb'))
dt3 = pickle.load(open('finalized_Knc_model1.sav', 'rb'))
```

```python
X_test = X_test[:, :1741]
```

```python
pred_df = pd.DataFrame()
pred_dt = dt1.predict(X_test)
pred_df['SVC'] = pred_dt
pred_rft = dt2.predict(X_test)
pred_df['RandomForest'] = pred_dt
pred_svm = dt3.predict(X_test)
pred_df['Kneighbors'] = pred_dt

df_ans = df_ans.drop(['text', 'text_clean'], axis = 1)
df_ans = df_ans.reset_index()
df_ans = df_ans.drop(['index'], axis = 1)
df_ans.head()
```

```python
df_ans['label'] = pd.Series(pred_dt)
df_ans.to_csv('tfidf_svc.csv', index=False)
df_ans.head(20)
```

```python
df_ans['label'] = pd.Series(pred_rft)
df_ans.to_csv('tfdif_rft.csv', index= False)
df_ans.head(20)
```

```python
df_ans['label'] = pd.Series(pred_svm)
df_ans.to_csv('tfdif_svm.csv', index= False)
df_ans.head(20)
```

```python
pred_df['res'] = pred_df.mode(axis = 1)[0].astype(int)
pred_df.head()
res = pred_df['res'].tolist()
```

```python
res
```

```python
res[0]
```

```python
label = []
for i in range(len(res)):
    ans = res[i]
    if (ans == 0):
        label.append('HATE')
    elif (ans == 1):
        label.append('NONE')
    elif (ans == 2):
        label.append('OFFN')
    elif (ans == 3):
        label.append('PRFN')
```

```python
label
```

```python
df_res = pd.DataFrame()
df_res['_id'] = df_ans['_id']
df_res['label'] = label
```

```python
df_res
```

```python
df_res.to_csv('final.csv', index=False)
```
